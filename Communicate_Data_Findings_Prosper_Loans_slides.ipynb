## Investigating Prosper Loans Datasets

### Loading Required Packages 

```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
sns.set(font_scale=1.2)
```

### Reading and Exploring Dataset

```python
df = pd.read_csv('prosperLoanData.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

### Data Preparation

```python
#converting dates into datetime objects 
df.ListingCreationDate = pd.to_datetime(df.ListingCreationDate)
df.LoanOriginationDate = pd.to_datetime(df.LoanOriginationDate)
df.ClosedDate = pd.to_datetime(df.ClosedDate)
```

```python
df.IncomeRange.unique()
df.IncomeRange[df.IncomeRange == '$25,000-49,999'] = 'Below 50K'
df.IncomeRange[df.IncomeRange == '$50,000-74,999'] = 'Below 75K'
df.IncomeRange[df.IncomeRange == '$75,000-99,999'] = 'Below 100K'
df.IncomeRange[df.IncomeRange == '$1-24,999'] = 'Below 25K'
df.IncomeRange[df.IncomeRange == '$100,000+'] = 'Above 100K'
df.IncomeRange[df.IncomeRange == '$0'] = 'No Income'
```

```python
df.LoanStatus [(df.LoanStatus == 'Past Due (1-15 days)' ) | (df.LoanStatus == 'Past Due (16-30 days)') | (df.LoanStatus == 'Past Due (31-60 days)' )] = 'past Due (<60 days)'
df.LoanStatus [(df.LoanStatus == 'Past Due (61-90 days)' ) | (df.LoanStatus == 'Past Due (91-120 days)')  ] = 'past Due (<120 days)'
```

```python
df.loc[df.Term == 36 , 'Term'] = '3 years'
df.loc[df.Term == 12 , 'Term'] = 'one year'
df.loc[df.Term == 60 , 'Term'] = '5 years'
```

```python
df.loc [df.CurrentlyInGroup == True , 'CurrentlyInGroup' ] = 'In a group'
df.loc [df.CurrentlyInGroup == False , 'CurrentlyInGroup' ] = 'Individual'
```

### Univariate Exploration

### What is the loan status of the majority of borrowers?

```python
plt.figure(figsize=(10,6))
sns.countplot(y="LoanStatus", data=df , color = sns.color_palette()[0]);
plt.xlabel('Loan Status')
plt.ylabel('Number of Listings')
plt.title('The current status of the loan'.title(),fontsize = 14, weight = "bold");
```

> the above chart shows that the loan status of the majority of loans is 'running loans' or 'completed'

### What is the income range of the majority of borrowers at the time the listings created?

```python
plt.figure(figsize=(10,6))
sns.countplot(y="IncomeRange", data=df , color = sns.color_palette()[0]);
plt.ylabel('Income Range')
plt.xlabel('Number of Listings')
plt.title('The income range of the borrower at the time the listing was created'.title(),fontsize = 14, weight = "bold");
```

```python
df['IncomeVerifiable'].value_counts(normalize = True)
```

> The income range of borrowers shows that the majority of borrowers' income is less than 75,000$, and around 92 percent provided the documents that supported and verfied their income

### What is the debt to income ratio of the majority of listings ?

```python
#univariate plot of debt income ratio on a standard scale
plt.figure(figsize=[10, 5])
binsize = 0.05
bins = np.arange(0, df ['DebtToIncomeRatio'].max()+binsize, binsize)
plt.hist(data = df, x = 'DebtToIncomeRatio', bins = bins)
plt.xlim([0, 1.3])
plt.xlabel('Debt To Income Ratio')
plt.ylabel('Number of Listings')
plt.title('Debt to Income Ratio'.title(),fontsize = 14, weight = "bold")
plt.show()
```

> Debt to income ratio for the majority of listings was less than 0.4, which means that the maximum debt amount that most of borrowers had obtianed was less than 30,000$

### What is the interest rate of the majority of listings ? 

```python
#univariate plot of borrower rate on a standard scale
plt.figure(figsize=[10, 5])
binsize = 0.05
bins = np.arange(0, df ['BorrowerRate'].max()+binsize, binsize)
plt.hist(data = df, x = 'BorrowerRate', bins = bins)
plt.xlim([0, 0.45])
plt.xlabel('Interset Rate')
plt.ylabel('Number of Listings')
plt.title('The Interest Rate of Listings'.title(),fontsize = 14, weight = "bold")
plt.show()
```

> the interest rate for the majority of loans was between 10 percent and 20 percent  

### Bivariate Exploration

#### In this part of the analysis I would like to investigate the correlation between some continous variables using the heatmap

```python
numeric_var = ['BorrowerRate' , 'LoanOriginalAmount' , 'DebtToIncomeRatio' , 'Recommendations' , 'StatedMonthlyIncome' , 'Investors']
```

```python
plt.figure(figsize = [10, 5])
sns.heatmap(df[numeric_var].corr(), annot = True, fmt = '.3f',
           cmap = 'vlag_r', center = 0)

plt.title('Correlation Matrix'.title(),fontsize = 14, weight = "bold")
plt.show()
```

```python
plt.figure(figsize = [10, 8])

sns.scatterplot(data=df, x="LoanOriginalAmount", y="Investors" );
plt.title('Loan amount and number of investors'.title(),fontsize = 14, weight = "bold");
```

```python
plt.figure(figsize = [10, 8])

sns.scatterplot(data=df, x="BorrowerRate", y="Investors" );

plt.title('Interest rate and number of investors'.title(),fontsize = 14, weight = "bold");
```

> the heatmap and scatter plots demonstrates that the loan amount is positively correlated the most with the borrower's monthly income and the number of investors that funded the loan, while it is negatively correlated with the interest rate. Hence, the interst rate is also negatively correlated with the number of investors that funded the loan.

### What is the debt to income ratio and interest rate of borrowers according to the loan status ?

```python
df.LoanStatus [(df.LoanStatus == 'Past Due (1-15 days)' ) | (df.LoanStatus == 'Past Due (16-30 days)') | (df.LoanStatus == 'Past Due (31-60 days)' )] = 'past Due (<60 days)'
df.LoanStatus [(df.LoanStatus == 'Past Due (61-90 days)' ) | (df.LoanStatus == 'Past Due (91-120 days)')  ] = 'past Due (<120 days)'
```

```python
df.LoanStatus.value_counts()
```

```python
plt.figure(figsize=[10, 8])
LoanStatus_incomeratio = df.groupby ("LoanStatus")['DebtToIncomeRatio'].agg(["mean"])
LoanStatus_incomeratio = LoanStatus_incomeratio.reset_index()
sns.barplot(x="mean", y="LoanStatus", data=LoanStatus_incomeratio , color = sns.color_palette()[1]);
plt.xlabel('Debt to Income Ratio');
plt.ylabel('Loan Status');
plt.title('Loan Status and Debt to income ratio'.title(),fontsize = 14, weight = "bold");
```

```python
plt.figure(figsize=[10, 8])
LoanStatus_borrowerrate = df.groupby ("LoanStatus")['BorrowerRate'].agg(["mean"])
LoanStatus_borrowerrate = LoanStatus_borrowerrate.reset_index()
sns.barplot(x="mean", y="LoanStatus", data=LoanStatus_borrowerrate , color = sns.color_palette()[3]);
plt.xlabel('Interest Rate');
plt.ylabel('Loan Status');
plt.title('Loan Status and Interest Rate'.title(),fontsize = 14, weight = "bold");
```

> the above to bar charts show that the debt to income ratio and interest rate for borrowers with loan status (Defaulted) or (Charged off) was higher than others completed their loans 

### Multivariate Exploration 

```python
# plt.figure(figsize=[10, ])
df_sub = df[['LoanStatus' , 'IncomeRange' , 'DebtToIncomeRatio' , 'BorrowerRate']]
df_sub_agg = df_sub.groupby(['LoanStatus' , 'IncomeRange'])['DebtToIncomeRatio'].agg(['mean'])
df_sub_agg = df_sub_agg.reset_index()
df_sub_agg
```

```python
plt.figure(figsize=[15, 10])
sns.barplot(x="mean", y="LoanStatus", data=df_sub_agg ,  hue= "IncomeRange" , color = sns.color_palette()[3]);
plt.legend(loc='lower right')
plt.ylabel('Loan Status');
plt.xlabel('Interest Rate');
plt.title('Income Range and Loan Status'.title(),fontsize = 14, weight = "bold");
```

> the above chart shows that the interest rate was the highest among those had defaulted or charged off loans and had no income or not employed 

#### In this part of the analysis, I aim to explore whether the loan amount is associated with the interest rate according to the length of the loan 

```python
sns.set(font_scale=1.5)
g = sns.FacetGrid(df, col="Term" , height=6, aspect= 0.8)
g.map(sns.scatterplot, "LoanOriginalAmount", "BorrowerRate");
g.set_axis_labels("Loan amount", "Interest rate")
#plt.title('Loan amount and Interest rate according to loan length'.title(),fontsize = 14, weight = "bold")
```

> the chart above shows that borrowers that received a short-term loan (one year) had relatively lower interset rate comparing with the 3years and 5 years loans, in addition their loan amount was less than 10,000 in most of listings.

> Borrowers received 3 years or 5 years loans, had the chance to obtain a higher loan amount ( up to 30,000 and more).

> the interest rate for the same loan amount was higher for long-term loans (3 years or 5 years) comparing with short-term loans

> In all cases, regardless the length of the loan, the higher amount obtained, the lower interest rate associated.

### Are the Interest rate and the loan amount that funded the loan associated according to the borrower income and the loan length?

```python
plt.figure(figsize=[20, 10])
g = sns.FacetGrid(df, col="IncomeRange" , col_wrap= 3 ,hue = 'Term' ,height=5, aspect=0.8 )
g.map(sns.scatterplot, "LoanOriginalAmount", "BorrowerRate");
g.set_axis_labels("Loan amount", "Interest rate");
g.add_legend();
```

> When the borrower's income is higher than 100K, the interest rate recorded  lower values for the long term loans (3 years or more) with an amount higher than 30000.

> When the borrower didn't provide any income proof, there the interest rate recorded higher values for all loans amounts.

### Conclusion 

> 1.  the loan amount is associated with the number of investors who fund the loan, the more investors the higher amount provided, in addition, the more investors contributed the lower interest rate the borrower will sustain.

> 2. the majority of borrower have to submit documents that verfies their income, which should not be less than $75,000 in average.

> 3. the debt to income ratio and interest rate for borrowers with loan status (Defaulted) or (Charged off) was higher than others completed their loans 

> 4. borrowers that received a short-term loan (one year) had relatively lower interset rate comparing with the 3years and 5 years loans, in addition their loan amount was less than 10,000 in most of listings.
